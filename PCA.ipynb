```python
data.drop(['education'],axis=1,inplace=True)
data.dropna(inplace=True)
```

```python
cov_mat = np.cov(data.T)
```

```python
print(cov_mat)
```

```python
eigval,eigvec = np.linalg.eig(cov_mat)
```

```python
eigval
```

```python
print(np.cumsum([i*(100/sum(eigval)) for i in eigval]))
```

```python
PC = eigvec.T[0:5]
```

```python
PC.T
```

```python
data_transformed = np.dot(data,PC.T)
```

```python
data_transformed.shape
```

```python
PCA_model = PCA(n_components=5,random_state=42)
```

```python
data_transformed1 = PCA_model.fit_transform(data)
```

```python
data_transformed1.shape
```

```python
data_transformed = pd.DataFrame(data_transformed)
data_transformed1 = pd.DataFrame(data_transformed1)
```
